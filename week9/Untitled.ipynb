```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,4*np.pi,360)
sin_vals = np.sin(x)

colors = ['b','g','r','o','m','k']

for order in range(3, 7):
    poly = np.polyfit(x, sin_vals, order)
    points = np.polyval(poly, x)
    plt.plot(x, points, label=order)    
plt.plot(x, sin_vals, label="original")
plt.legend()
plt.show()
```
